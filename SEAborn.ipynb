```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd
%matplotlib inline
```

```python
# Read dataset
df = pd.read_csv('Pokemon.csv', index_col=0)
```

```python
df.head()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df)

# sns.lmplot(x=df.Attack, y=df.Defense)
```

```python
# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,fit_reg=False,hue='Generation')
```

```python
# Boxplot
sns.boxplot(data=df)
```

```python
#Set theme
sns.set_style('whitegrid')
 
# Violin plot
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
# Violin plot with palette
sns.violinplot(x='Type 1', y='Attack', data=df,   palette='winter_r') # Set color palette
```

```python
# Violin plot with  color
sns.violinplot(x='Type 1', y='Attack', data=df, color='red')
```

```python
# Swarm plot with palette
sns.swarmplot(x='Type 1', y='Attack', data=df,  palette='winter_r')
```

```python
# Set figure size with matplotlib
plt.figure(figsize=(20,16))
 
# Create plot
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette='spring')
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Attack by Type')
```
